```python
import requests
import io
from PIL import Image
import json
```

```python
with open('cred.json') as credentials:
    headers = json.load(credentials)
```

```python
# API_URL = "https://api-inference.huggingface.co/models/CompVis/stable-diffusion-v1-4" #Testar com esse modelo tamb√©m
API_URL = "https://api-inference.huggingface.co/models/anant9/website-ui-sdxl-lora"
# headers = {"Authorization": "Bearer hf_xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"} #Adicionar o token criado da HuggingFace para acessar a API
```

```python
def query(payload):
	response = requests.post(API_URL, headers=headers, json=payload)
	return response.content
```

```python
image_bytes = query({
	"inputs": "A website ui screenshot for a donuts business",
})

image = Image.open(io.BytesIO(image_bytes))
```
